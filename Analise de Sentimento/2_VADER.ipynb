### Polarity Score
- Tipo dicionario (chave: valor)
- Negativo, Neutro, Positivo ou Compound
- Compound: valor real [-1, 1] muito negativo ou muito positivo
  - Calculada atraves das palavras de sentença

---

### Textblob
- Polarity: valor real [-1, 1] muito negativo ou muito positivo
- Subjectivity: O quanto a sentença se refere a opiniao, julgamento ou emoçao
  - Valor real [0,1]. quanto mais proximo de 1, mais subjetivo

# Aula 2: VADER na pratica

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
instancia = SentimentIntensityAnalyzer()

a = 'I HATE this movie!!'
x = instancia.polarity_scores(a)
print(x)
```

```python
a = 'This plot is very old'
x = instancia.polarity_scores(a)
print(x)
```

```python
a = ':)'
x = instancia.polarity_scores(a)
print(x)
```

```python
a = ':('
x = instancia.polarity_scores(a)
print(x)
```

```python
a = 'I HATE this movie! But I loved the girl'
x = instancia.polarity_scores(a)
print(x)
```

```python
print(type(x))
print(x['neg'])
print(x['compound'])
```

```python
from textblob import TextBlob
```

```python
teste = TextBlob("The movie was awesome!")
print(teste.sentiment)
```

```python
teste = TextBlob("I love NY!")
print(teste.sentiment)
```
