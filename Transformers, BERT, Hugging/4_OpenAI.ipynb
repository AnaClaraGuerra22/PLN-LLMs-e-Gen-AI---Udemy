```python
response = client.chat.completions.create(
    model=model,
    messages=[
        { "role": "system",
         "content": "You are a investment assistant."
         },

         { "role": "user",
          "content": "What is the best low-risk investment you recommend for this year?"
        }
    ]
)
```
