# Aula 3: Aplicação de Perguntas e Respostas ( question-answering )

```python
# https://huggingface.co/pierreguillou/bert-base-cased-squad-v1.1-portuguese

qea = pipeline("question-answering", model= "pierreguillou/bert-base-cased-squad-v1.1-portuguese")
```

```python
texto = "Ediscere, Scire, Agere, Vincere (Aprender, Saber, Agir, Vencer) frase que ressalta sua excelência em educação e pesquisa, como referência na educação e excelência do ensino, pesquisa e extensão, e expressões que demonstram a paixão e o orgulho dos estudantes por pertencerem à instituição, como UFV significa Universo de possibilidades, Futuro promissor, Vencer por meio de esforço e dedicação "

pergunta = "O que a frase ressalta?"
#pergunta = "a UFV é referencia de que?"

resposta = qea(context=texto, question=pergunta)

print("Pergunta: ", pergunta)
print("Resposta: ", resposta["answer"])
print("Score: ", resposta["score"])
```

# Aula 4: Aplicação de Preenchimento de Lacunas (Fill - Mask)

https://huggingface.co/neuralmind/bert-base-portuguese-cased



```python
mascarar = pipeline("fill-mask", model = "neuralmind/bert-base-portuguese-cased")
```

```python
# quero que ele preveja 'chao'
texto = mascarar("Existe uma chance do copo cair no [MASK]")

for x in range(len(texto)):
  print(texto[x])
```

```python
# quero que ele preveja 'chao' e 'Brasil'

texto = mascarar(["Existe uma chance do copo cair no [MASK]", "Brasília é a capital do [MASK]"])

for x in range(len(texto)):
  print(texto[x])
```

# Aula 5: Aplicação de Resumos ( summarization )

```python
resumir = pipeline("summarization")
```

```python
texto = """Modelos de linguagem de grande escala (em inglês: Large Language Model ou LLM)[1][2] são modelos de linguagem compostos por uma rede neural com muitos parâmetros (tipicamente bilhões ou possivelmente mais[3]). São treinados com grandes quantidades de textos não rotulado usando aprendizado de máquina não-supervisionado.[4][5] Os LLM surgiram por volta de 2018,[6] com o modelo BERT.[6] Estes têm bom desempenho em uma ampla variedade de tarefas. Isso mudou o foco da pesquisa em processamento de linguagem natural, afastando-se do paradigma anterior de treinar modelos supervisionados especializados para tarefas específicas. Propriedades: Embora o termo não tenha uma definição formal, ele geralmente se refere a modelos de aprendizado profundo que possuem uma contagem de parâmetros da ordem de bilhões ou mais.[7] MLLs são modelos de propósito geral que se destacam em uma ampla gama de tarefas, em vez de serem treinados para uma tarefa específica (como análise de sentimento, reconhecimento de entidades nomeadas ou raciocínio matemático). Embora treinados em tarefas simples, como prever a próxima palavra em uma frase, modelos de linguagem neural com treinamento e contagem de parâmetros suficientes são capazes de capturar grande parte da sintaxe e semântica da linguagem humana. Além disso, grandes modelos de linguagem demonstram considerável conhecimento geral sobre o mundo e são capazes de "memorizar" uma grande quantidade de fatos durante o treinamento."""
```

```python
resumo = resumir(texto, max_length = 100, min_length=50)
print(resumo)
```

# Aula 6: Aplicação de Geração de Texto ( Text Gerenation )

```python
# https://huggingface.co/pierreguillou/gpt2-small-portuguese

gerar = pipeline("text-generation", model = "pierreguillou/gpt2-small-portuguese")
```

```python
texto = "Em sentido estrito, a ciência refere-se ao sistema de adquirir conhecimento baseado no método cientifíco."
resultado = gerar(texto, max_length=60, do_sample=True)
print(resultado)
```
