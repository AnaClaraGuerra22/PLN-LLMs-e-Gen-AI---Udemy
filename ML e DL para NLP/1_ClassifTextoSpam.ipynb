# Aula 2: Classificação de Spam I

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
url = "https://raw.githubusercontent.com/AnaClaraGuerra22/PLN-LLMs-e-Gen-AI---Udemy/52804be1d038ac399f9c9b3594330a5df38f808c/ML%20e%20DL%20para%20NLP/spam.csv"

spam = pd.read_csv(url)
print(spam.shape)
spam.head()
```

```python
spam['Category'].value_counts()

# desbalanceado, tem muito mais Ham do que Spam
```

```python
previ = spam['Message']
classe = spam['Category']

#  TfidfVectorizer atribui pesos as mensagens (faz matrizes nas palavras)

vetorizador = TfidfVectorizer()
previsores = vetorizador.fit_transform(previ)

previsores.shape
```

```python
print(vetorizador.get_feature_names_out()[10:100])
```

# Aula 3: Classificação de Spam II

```python
# divisao de treino e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(previsores, classe, test_size=0.3)
```

```python
x_teste.shape
```

```python
# random forest

florest = RandomForestClassifier(n_estimators=500)
florest.fit(x_treino, y_treino)
```

```python
previsoes = florest.predict(x_teste)

print(confusion_matrix(y_teste, previsoes))

# nao tem nenhum falso positivo de spam, isso é bom (menos chance de perder mensagens importantes)
```

```python
print(accuracy_score(y_teste, previsoes))
print(metrics.classification_report(y_teste, previsoes))
```
